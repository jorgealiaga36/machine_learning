# Matplotlib 

```python
from matplotlib import pyplot as plt
%matplotlib inline

import pandas as pd
import numpy as np
```

```python
X = np.linspace(-np.pi, np.pi, 256)
S, C = np.sin(X), np.cos(X)
```

```python
#Plotting with default settings (at a time).

plt.plot(X, C)
plt.plot(X, S)

plt.show()
```

```python
#Plotting with default settings (separately).

#Charts are showed into 2 rows and 1 column (Number os axes).

plt.subplot(2, 1, 1)  #It is plotted the first chart (third argument).
plt.plot(X, C)

plt.subplot(2, 1, 2) #It is plotted the second chart (third argument).
plt.plot(X, S)

plt.show()
```

```python
#Plotting with default settings (separately).

#Charts are showed into 2 rows and 2 columns.

plt.subplot(2, 2, 1)  #It is plotted the first chart (third argument).
plt.plot(X, C)

plt.subplot(2, 2, 2) #It is plotted the second chart (third argument).
plt.plot(X, S)

plt.subplot(2, 2, 3)  #It is plotted the third chart (third argument).
plt.plot(2*X, C)

plt.subplot(2, 2, 4) #It is plotted the fourth chart (third argument).
plt.plot(2*X, S)

plt.show()
```

```python
#Plotting instantiating defaults (at a time). Setting x and y limits / x and y ticks.

plt.figure(figsize = (5,4), dpi = 80) #Create a figure of size 8x6 inches, 80 dots per inch (resolution).

plt.plot(X, C, color = "blue", linewidth = 1.0, linestyle = '-', label = 'sin(X)')    #Plot cosine with a blue continuous 
                                                                                      #line of width 1 (pixels).
plt.plot(X, S, color = "green", linewidth = 1.0, linestyle = '--', label = 'cos(X)')

plt.xlim(-4, 4) #Set x limits.
plt.xticks(np.linspace(-4, 4 , 9)) #Set x ticks (separations).

plt.ylim(-1, 1)
plt.yticks(np.linspace(-1, 1 , 5))

plt.legend()

plt.show()
```

```python
#Plotting instantiating defaults (separately and at a time). Setting x and y limits / x and y ticks correctly.

plt.figure(figsize = (6,7), dpi = 80) 

plt.subplot(2,1,1)

plt.plot(X, C, color = "blue", linewidth = 1.0, linestyle = '-', label = 'cos(X)') 
plt.plot(X, S, color = "green", linewidth = 1.0, linestyle = '-', label = 'sin(X)')

plt.xlim(X.min() * 1.05, X.max() * 1.05) #Making sure of the function's limits. Extra margin to ensure the function 
                                         #is not too tight plotted.
plt.xticks(np.linspace(-3, 3, 7))
plt.ylim(C.min() * 1.1, C.max() * 1.1)
plt.yticks(np.linspace(-1, 1 , 5))

plt.legend(loc = 'upper left') #upper/lower right/left.


plt.subplot(2,1,2)

plt.plot(2*X, C, color = "red", linewidth = 1.0, linestyle = '-', label = 'cos(2X)' ) 
plt.plot(2*X, S, color = "grey", linewidth = 1.0, linestyle = '-', label = 'sin(2X)')

plt.xlim(X.min() * 1.1, X.max() * 1.1) 
plt.xticks(np.linspace(-4, 4 , 9))
plt.ylim(C.min() * 1.1, C.max() * 1.1)
plt.yticks(np.linspace(-1, 1 , 5))

plt.legend(loc = 'best') #The best location for the leyend in the subplot.


plt.show()
```

```python
#Plotting setting tick labels. Using latex for nice rendering.

plt.figure(figsize = (4,4), dpi = 80)

plt.plot(X, C, linewidth = 1.0, color = 'red', linestyle = '-')

plt.xlim(X.min() * 1.1, X.max() * 1.1)
plt.ylim(C.min() * 1.1, C.max() * 1.1)

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

plt.yticks([-1, 0, 1], [r'$-1$', r'$0$', r'$1$'])

plt.show()
```
